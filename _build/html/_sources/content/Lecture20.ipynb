## 20.3 Scanning Tunneling Microscopy (STM)

### 20.3.1 Principle:
- A sharp **metal tip** is brought close to a conducting surface.
- A **voltage** is applied between tip and sample.
- **Quantum tunneling** of electrons occurs:
  - When the tip is near the surface, electrons can "tunnel" through the vacuum gap.
  - **Tunneling current** increases as tip gets closer.

### 20.3.2 Imaging:
- If current is kept constant, the tip **moves up and down** to maintain it.
- This motion is used to **map the surface** with atomic resolution.

> ðŸ’¡ STM was the first technique to directly **image individual atoms** on surfaces.

## 20.4 Atomic Force Microscopy (AFM)

### 20.4.1 How it works:
- A **cantilever with a sharp tip** is brought near the sample surface.
- **Forces** between the tip and surface (van der Waals, electrostatic, etc.) cause the cantilever to deflect.
- A **laser beam** tracks the deflection.

### 20.4.2 Hookeâ€™s Law:

$$F = kx$$
- **F**: Force
- **k**: Spring constant of cantilever
- **x**: Deflection

### 20.4.3 Advantages:
- Works on **non-conductive materials**
- Can image **soft biological samples** (e.g. DNA, proteins)
- Can image in aqueous environment

## 20.5 Fluorescence Microscopy

- Used to image **specific molecules** in biological samples.
- Sample is treated with **fluorescent dyes** or tagged proteins.
- When excited by light of one wavelength, fluorophores **emit light** at another wavelength.

> âœ… Useful for **live-cell imaging**, tracking molecules, and bio-labeling.
