# 9. Carbon Materials


## 9.1 Hybridisation of s and p Orbitals 

- **Hybridisation** occurs when atomic orbitals mix to form new, equivalent hybrid orbitals.  
- **sp Hybridisation:**  
  - The **2s orbital** mixes with just **one** of the **2p orbitals** to form **two equivalent sp orbitals**.
  - The two sp orbitals are **arranged linearly**, with the remaining **2p orbitals perpendicular** to them.

   
### 9.1.1 sp¹ Hybridized Carbon & Bonding in Ethyne (C₂H₂)
- In ethyne (acetylene), **each carbon is sp hybridized**, leading to:
  1. Two **C-H σ bonds**:
     
     $$C(sp^1) - H(1s)$$
  2. One **σ bond** between the two carbons:
     
     $$C(sp^1) - C(sp^1)$$
  3. Two **π bonds** (overlapping p orbitals), creating a **triple bond (C≡C)**:
     
     $$C \equiv C$$

## 9.2 Conjugation and Delocalisation

- **Conjugation** occurs when **double and single bonds alternate**, allowing **p orbitals to overlap**, forming a **delocalized π electron cloud**.
- **Effects of Conjugation:**
  - Leads to **characteristic spectral absorptions** (UV-visible).
  - As conjugation increases, the energy difference between bonding and anti-bonding π orbitals **decreases**, causing **absorption in the visible range**.

### 9.2.1 Delocalisation and Aromaticity
- **Aromaticity** occurs when cyclic compounds have **delocalized π molecular orbitals** extending over all atoms.
- **Example: Benzene (C₆H₆)**

# 9.3 Carbon, Nitrogen, and Oxygen: Electron Configuration and Lone Pairs

## 9.3.1 Electron Configurations

- **Carbon (C):**  
  $$1s^2 \ 2s^2 \ 2p^2$$

- **Nitrogen (N):**  
  $$1s^2 \ 2s^2 \ 2p^3$$

- **Oxygen (O):**  
  $$1s^2 \ 2s^2 \ 2p^4$$

These atoms form covalent bonds by hybridising their valence orbitals into **sp³ hybrid orbitals**, which can accommodate **bonding pairs** or **lone pairs** of electrons.

---

## 9.3.2 Orbital Diagrams of sp³ Hybridisation

### Carbon in CH₄ (Methane): 4 valence electrons → 4 bonding pairs

$$
\begin{array}{cccc}
\text{sp}^3 & \text{sp}^3 & \text{sp}^3 & \text{sp}^3 \\
\uparrow & \uparrow & \uparrow & \uparrow \\
\end{array}
$$

→ All orbitals contain one electron — ready to form **4 sigma bonds**.

---

### Nitrogen in NH₃ (Ammonia): 5 valence electrons → 3 bonding + 1 lone pair

$$
\begin{array}{cccc}
\text{sp}^3 & \text{sp}^3 & \text{sp}^3 & \text{sp}^3 \\
\uparrow & \uparrow & \uparrow & \uparrow\downarrow \\
\end{array}
$$

→ 3 orbitals form N–H bonds; the fully occupied one is a **lone pair**.

---

### Oxygen in H₂O (Water): 6 valence electrons → 2 bonding + 2 lone pairs

$$
\begin{array}{cccc}
\text{sp}^3 & \text{sp}^3 & \text{sp}^3 & \text{sp}^3 \\
\uparrow\downarrow & \uparrow & \uparrow\downarrow & \uparrow \\
\end{array}
$$

→ Two lone pairs, two bonding pairs → **bent** geometry.

---

## 9.3.3 Molecular Shapes and Bond Angles

| Molecule | Shape               | Bond Angle | Lone Pairs |
|----------|---------------------|------------|------------|
| CH₄      | Tetrahedral         | 109.5°     | 0          |
| NH₃      | Trigonal pyramidal  | ~107°      | 1          |
| H₂O      | Bent (V-shaped)     | ~104.4°    | 2          |

> **Lone pairs** cause stronger repulsion than bonding pairs, which compresses the bond angles.

---

## 9.3.4 Summary

- **Lone pairs occupy sp³ hybrid orbitals** and repel more strongly than bonded electrons.
- This causes **distortions in bond angles** and leads to **different molecular geometries**.
- Electron occupancy diagrams help visualise how hybrid orbitals are filled and used.

### 9.4.2 Van der Waals Interactions
- **Van der Waals forces** arise from **fluctuating dipoles** that create weak **attractive interactions**.
- **Factors influencing strength:**
  - **Size of electron cloud** (larger atoms/molecules have stronger Van der Waals forces).
  - **Proximity of molecules**.

### 9.4.3 Lennard-Jones Potential (LJ Potential)
- Describes **Van der Waals interactions** mathematically:
  - **Attractive term** (dominant at longer distances).
  - **Repulsive term** (dominant at short distances, due to the Pauli exclusion principle).
  - The **Lennard-Jones potential** equation:
    
    $$U(r) = 4\epsilon \left[ \left(\frac{\sigma}{r}\right)^{12} - \left(\frac{\sigma}{r}\right)^6 \right]$$
    where:
    - $\epsilon $ = depth of the potential well (interaction strength).
    - $ \sigma $ = atomic radius.
    - $ r $ = interatomic separation.

## 9.5 Carbon Allotropes
- **Carbon exists in different structural forms** due to its ability to form multiple bonding types.  Theer are three allotropes of carbon/

### 9.5.1 Graphite

#### 9.5.1.1 Physical Properties:
- **Soft and Slippery:** Layers of carbon atoms are held together by weak van der Waals forces, allowing them to slide over each other.
- **Grey-Black, Opaque Appearance:** It has a metallic sheen but is not reflective like metals.
- **Lightweight:** Density of ~2.26 g/cm³.
- **Good Electrical Conductor:** Each carbon atom is bonded to three others, leaving a free electron that moves through the layers.
- **High Melting Point (~3,600°C):** Strong covalent bonds within layers require significant energy to break.
- **Insoluble in Water and Organic Solvents:** Strong covalent bonds prevent dissolution.
- **Structure:**  
  - **Layers of carbon atoms** arranged in **hexagonal sheets**.
  - **Each carbon is sp² hybridized**, forming **σ and π bonds**.
  - **Weak interlayer forces** allow easy sliding between layers.
- **Properties:**
  - **Good electrical conductor** (due to delocalized electrons).
  - **Good conductor of heat**
  - **High thermal stability**.
  - **Self-lubricating** (low friction).
- **Applications:**  
  - **Pencils**, **electrodes**, **brake drums**.
#### 9.5.1.2 Key Equations:
- Electrical Conductivity: 
 
  $\sigma = \frac{1}{\rho}$
 
  where $ \rho $ is the electrical resistivity.
- Thermal Conductivity: 
  
  $k = \frac{1}{3} C_v v l$
 
  where $ C_v $ is the heat capacity, $ v$ is the speed of sound in the material, and $ l $ is the mean free path of phonons.

#### 9.5.1.3 Example Calculation:
**Estimate the thermal conductivity of graphite given:**
$ C_v = 2.0 \times 10^6 $ J/m³K,$ v = 2000 $ m/s, and $ l = 10^{-7} $ m.

**Solution:**

$k = \frac{1}{3} (2.0 \times 10^6) (2000) (10^{-7})$

$k = \frac{1}{3} (4.0 \times 10^5)$

$k = 1.33 \times 10^5 \text{ W/mK}$

### 9.5.2 Diamond
#### 9.5.2.1 Physical Properties:
- **Extremely Hard:** Strong covalent bonding in a tetrahedral structure makes it the hardest natural material.
- **Transparent and Colourless:** Impurities can add color (e.g., nitrogen makes it yellow, boron makes it blue).
- **High Density (~3.51 g/cm³).**
- **Poor Electrical Conductor:** All four valence electrons of carbon are bonded, leaving no free electrons.
- **Excellent Thermal Conductor:** Better than metals like copper.

#### 9.5.2.2 Key Equations:
- Young’s Modulus: 
  
  $E = \frac{\sigma}{\varepsilon}$
  
  where $ \sigma $ is the applied stress and $ \varepsilon $ is the strain.

- **Structure:**
  - **Each carbon forms four σ bonds**, creating a **3D tetrahedral network**.
  - **Strong covalent bonds** make it extremely hard.
- **Properties:**
  - **Transparent** and **colorless**.
  - **Hardest naturally occurring material**.
  - **Electrical insulator** but **excellent heat conductor**.
- **Applications:**  
  - **Cutting tools**, **jewelry**, **drill tips**.

### 9.5.3 Comparison of Diamond and Graphite

| **Property**            | **Diamond**                                      | **Graphite**                                    |
|-------------------------|------------------------------------------------|------------------------------------------------|
| **Structure**           | Tetrahedral, 3D network of carbon atoms       | Hexagonal layers of carbon atoms              |
| **Bonding**            | Each carbon bonded to 4 others (strong covalent bonds) | Each carbon bonded to 3 others (weak van der Waals forces between layers) |
| **Hardness**           | Extremely hard (hardest natural material)     | Soft and slippery (layers slide over each other) |
| **Density**            | High (~3.51 g/cm³)                            | Lower (~2.26 g/cm³)                           |
| **Appearance**         | Transparent, colorless (pure diamond)         | Opaque, grey-black with a metallic sheen     |
| **Electrical Conductivity** | Does not conduct electricity (no free electrons) | Good conductor (delocalized electrons move between layers) |
| **Thermal Conductivity** | Excellent thermal conductor                 | Also a good thermal conductor                 |
| **Melting Point**      | Very high (~3,600°C)                          | Very high (~3,600°C)                          |
| **Solubility**         | Insoluble in water and organic solvents       | Insoluble in water and organic solvents       |
| **Reactivity**         | Chemically inert, burns in oxygen at ~850-1000°C | Chemically stable, oxidizes at high temperatures |
| **Uses**               | Jewelry, cutting tools, electronics, abrasives | Pencils, lubricants, batteries, electrodes, nuclear reactors |
