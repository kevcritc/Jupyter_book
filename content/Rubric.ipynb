# 21. Nanotechnology Essay Rubric

## Summary Intro (10 marks)

| Level | Description |
|---|---|
| Poor | - No summary or irrelevant content. <br> - Lacks structure or is repetitive. <br> - Insufficient detail to understand the essay. |
| Limited | - Attempts a summary but lacks sufficient detail. <br> - Weak structure. <br> - Limited engagement for the reader. |
| Adequate | - Covers main points but lacks depth or clarity. <br> - Logical structure but not engaging. |
| Good | - Concise, well-structured introduction. <br> - Clearly indicates essay content. <br> - Engages the reader. |
| Excellent | - Exceptionally clear and structured summary (<300 words). <br> - Highly engaging and insightful. <br> - Provides a comprehensive overview. |
| Outstanding | Outstanding |

---

## Background (25 marks)

| Level | Description |
|---|---|
| Poor | - Fails to introduce the science. <br> - Serious scientific errors. <br> - Information is irrelevant or poorly structured. |
| Limited | - Introduces background but with minor flaws. <br> - Lacks depth and supporting evidence. <br> - Weak structure. |
| Adequate | - Provides relevant background but includes trivial points. <br> - Minor inaccuracies or unclear explanations. |
| Good | - Detailed background with logical structure. <br> - Scientifically accurate and well-supported. |
| Excellent | - Comprehensive and scientifically accurate background. <br> - Uses examples effectively. <br> - Logical, layered structure with excellent clarity. |
| Outstanding | Outstanding |

---

## Discussion (25 marks)

| Level | Description |
|---|---|
| Poor | - Lacks critical analysis. <br> - Irrelevant or unsupported points. <br> - No clear link to the essay topic. |
| Limited | - Discusses aspects but with weak assessment. <br> - Limited connection to the main topic. |
| Adequate | - Covers application and impact but lacks depth. <br> - Discussion is present but not comprehensive. |
| Good | - Well-structured, critical discussion of impact. <br> - Evidence-based analysis. <br> - Addresses challenges effectively. |
| Excellent | - Outstanding critical analysis. <br> - Balanced, evidence-backed arguments. <br> - Introduces new perspectives and insights. |
| Outstanding | Outstanding |

---

## Style (10 marks)

| Level | Description |
|---|---|
| Poor | - No title, figures, or captions. <br> - Poor format and structure. <br> - Frequent grammar errors. |
| Limited | - Title present but weak. <br> - Figures included but lack relevance or captions. <br> - Errors obscure meaning at times. |
| Adequate | - Figures present but inconsistently useful. <br> - Adequate structure but lacks polish. <br> - Generally clear language with minor errors. |
| Good | - Relevant figures referenced in text. <br> - Clear structure and professional formatting. <br> - Minor grammar issues. |
| Excellent | - Excellent formatting with professional structure. <br> - Two relevant figures, clear captions, and proper integration. <br> - Fluent, polished writing. |
| Outstanding | Outstanding |

---

## Conclusions (5 marks)

| Level | Description |
|---|---|
| Poor | - No conclusion or completely irrelevant. |
| Limited | - Attempts a summary but misses key points. |
| Adequate | - Covers key points but lacks depth or clarity. |
| Good | - Concisely summarizes main points with a clear take-home message. |
| Excellent | - Insightful conclusion summarizing all major points. <br> - Introduces new, relevant insights. |
| Outstanding | Outstanding |

---

## Research (25 marks)

| Level | Description |
|---|---|
| Poor | - No references provided. <br> - No engagement with academic sources. <br> - Information is unsupported or plagiarized. |
| Limited | - Minimal or low-quality references (e.g., blogs or personal opinions). <br> - Over-reliance on a single source. <br> - Limited engagement with academic research. |
| Adequate | - Some high-quality references used but reliance on weaker sources. <br> - Limited engagement with primary sources. <br> - References inconsistently cited. |
| Good | - Uses a range of high-quality references (peer-reviewed articles, books, reports). <br> - References well integrated and cited. <br> - Some evidence of critical engagement. |
| Excellent | - Extensive use of peer-reviewed sources (≥6 references with ≥4 books or journal articles). <br> - Sources critically assessed. <br> - Accurate and consistent citation style. |
| Outstanding | Outstanding |
