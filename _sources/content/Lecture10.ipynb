# 10 Introduction to Carbon Nanomaterials
Nanomaterials are materials that have structural features at the nanometer scale, typically below 100 nm. Carbon nanomaterials exhibit unique mechanical, electrical, and thermal properties, making them essential in modern technology.

### 10.1 Carbon Nanotubes (CNTs)

Carbon Nanotubes (CNTs) are cylindrical structures made entirely of **carbon atoms** arranged in a **hexagonal lattice**, similar to **graphene**. Imagine a sheet of graphene rolled into a seamless cylinder.

There are two main types:
- **Single-Walled Carbon Nanotubes (SWCNTs)** – made of one graphene cylinder.
- **Multi-Walled Carbon Nanotubes (MWCNTs)** – multiple concentric graphene cylinders.

---

### 10.1.1 Structure

- We can think of rolling up a sheet graphene, and the way it's rolled defines the properties of the CNT. In realisty this is not how CNTs are made - they are chemically synthesised and the starting material is not graphene!

### 10.1.2 Chirality and Types
The way the graphene is rolled is described by the **chiral vector** (**n, m**), which leads to different types of CNTs:
- **Armchair (n = m):** Metallic
- **Zigzag (m = 0 or n = 0):** Can be metallic or semiconducting
- **Chiral (n ≠ m):** Usually semiconducting

This impacts their **electronic properties**.

---

## 10.2 Physical Properties

### 10.2.1 Mechanical Strength
- CNTs are **extremely strong** – stronger than steel (by weight).
- High **Young’s modulus** (~1 TPa) – they’re very stiff.
- Very high **tensile strength** (~100 times that of steel).

### 10.2.2 Electrical Properties
- Can behave as **metals or semiconductors** depending on chirality.
- **Ballistic conduction**: electrons travel with minimal scattering over long distances.
- Potential use in **nanoscale transistors**, wires, and quantum devices.

### 10.2.3 Thermal Conductivity
- Excellent **thermal conductors** along the tube (~3000–6000 W/m·K).
- Poor heat conductors **across** the tube walls due to weak van der Waals forces.

### 10.2.4 Density and Size
- **Very low density**: ~1.3–1.4 g/cm³.
- Typical diameter: ~1 nm (SWCNT), up to 50 nm (MWCNT).
- Length: from micrometres up to centimetres.

### 10.2.6 Key Equations:
- Electrical Conductivity: 

  $\sigma = \frac{q n \mu}{m}$
  
where $ q $ is the charge, $ n $ is the carrier density, $ \mu $ is the mobility, and $ m $ is the effective mass.

### 10.2.7 Example Calculation:
**Find the electrical conductivity of a CNT if charge** $ q = 1.6 \times 10^{-19} $ **C, carrier density** $ n = 10^{28} $ **electrons/m³, mobility** $ \mu = 0.1 $ **m²/Vs, and effective mass** $ m = 9.1 \times 10^{-31} $ **kg.**

**Solution:**

$\sigma = \frac{(1.6 \times 10^{-19}) (10^{28}) (0.1)}{9.1 \times 10^{-31}}$

$\sigma = \frac{1.6 \times 10^8}{9.1 \times 10^{-31}}$

$\sigma = 1.76 \times 10^7 \text{ S/m}$

## 10.3 Graphene

### 10.3.1 Physical Properties:

- Two-Dimensional Structure: A single layer of carbon atoms arranged in a hexagonal lattice.

- High Electrical Conductivity: Due to delocalized π-electrons that move freely.

- High Mechanical Strength: Young’s modulus of ~1 TPa, higher than steel.

- High Thermal Conductivity: ~5000 W/mK, among the highest of known materials.

- Optical Transparency: Absorbs ~2.3% of incident light, making it useful in transparent electronic devices.

#### 10.4.1 Applications:

- Nanoelectronics: Graphene's high electrical conductivity and exceptional charge carrier mobility (~200,000 cm²/V·s) enable ultra-fast transistors and sensors. It is used in field-effect transistors (FETs) for high-performance computing and quantum devices.

- Flexible Electronics: Due to its thinness and mechanical strength, graphene is employed in flexible displays, foldable smartphones, and wearable technology, maintaining conductivity even under deformation.

- Energy Storage: Graphene-based supercapacitors and lithium-ion batteries offer high charge storage, rapid charge-discharge cycles, and enhanced lifespan due to its large surface area (~2630 m²/g).

- Biomedical Applications: Graphene’s biocompatibility, high surface area, and antibacterial properties make it suitable for biosensors, targeted drug delivery, and neural interfaces. Its exceptional electrical properties enhance electrode performance in medical devices.

- Water Filtration: Graphene oxide membranes selectively filter out salts and contaminants due to their atomic-scale precision, enabling advanced desalination and purification technologies.

- Heat Dissipation: Graphene’s ultrahigh thermal conductivity (~5000 W/mK) is leveraged in cooling solutions for microprocessors, LEDs, and high-power electronics, preventing overheating and enhancing performance.

- Composites & Coatings: Graphene is added to polymers, metals, and ceramics to enhance mechanical strength, electrical conductivity, and thermal stability, making it useful in aerospace, automotive, and construction materials.

- Optoelectronics: Graphene’s ability to absorb and manipulate light enables applications in photodetectors, transparent conductive coatings for touchscreens, and next-generation optical communication devices.
