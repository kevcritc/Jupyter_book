# 16. Nanotechnology & Biology
- Nanotechnology exists in living organisms.
- Biological molecular machinery follows a **"bottom-up"** approach.
- Inspired the design of **nanobiotechnology**.

---

## 16.1 Chemistry of Life

### 16.1.1 **Water**
- Polar molecule, essential for life.
- Solvent for biochemical reactions.
- Forms **hydrogen bonds** (up to 4).
- **Hydrophilic**: Attracted to water.
- **Hydrophobic effect**: Nonpolar molecules aggregate.

### 16.1.2 **Carbon**
- Forms chains, rings, and branched structures.
- Carbon-rich regions are **hydrophobic**.

---

## 16.2 Molecules & Building Blocks of Life

### 16.2.1 Carbohydrates (Sugars)
- **Formula:** (CH₂O)n.
- **Functions:**
  - **Energy source** (glucose via glycolysis → ATP).
  - **Structural role** (polysaccharides).
- **Types:**
  - **Monosaccharides**: Single unit (e.g., glucose).
  - **Disaccharides** → **Polysaccharides** (e.g., starch, glycogen, cellulose).
- **Solubility**: Water-soluble, crystalline.

---

### 16.2.2 Lipids (Fatty Acids & Phospholipids)
- Composed of **carbon, hydrogen, oxygen**.
- **Fatty acids**:
  - **Hydrophilic head**, **hydrophobic tail** → Forms **micelles** in water.
  - **Functions**:
    - Structural component in membranes (**lipid bilayers**).
    - Energy storage (**minor role**).
  - **Types**:
    - **Saturated** (no double bonds, solid at room temp).
    - **Unsaturated** (one or more double bonds, liquid at room temp).
    - **Trans fats** (see below).

#### 16.2.3 **Trans Fats – What Are They?**
- A type of **unsaturated fat** with a **trans configuration** at the double bond.
- **Two types**:
  - **Natural** (small amounts in dairy & meat from ruminants).
  - **Artificial** (made by **partial hydrogenation** of vegetable oils).
- **Health Effects**:
  - Increase **LDL (bad cholesterol)** and lower **HDL (good cholesterol)**.
  - Linked to **heart disease, stroke, and type 2 diabetes**.
  - Many countries have **banned or restricted** artificial trans fats.

---

### 16.2.4 Nucleotides & Nucleic Acids**
- **Main Roles**:
  - Store genetic information.
  - Carry chemical energy.
- **Structure**:
  - **Five-carbon sugar**.
  - **Phosphate group(s)**.
  - **Nitrogen base**.
- **Types**:
  - **DNA (Deoxyribonucleic acid)**.
  - **RNA (Ribonucleic acid)**.
- **ATP (Adenosine Triphosphate)**:
  - "Energy currency" of cells.
  - Energy stored in phosphate bonds.
  - **ATP → ADP** releases **7 kcal/mol** of energy.

---

### 16.2.5 Amino Acids & Proteins
- **Structure**:
  - Central **α-carbon**.
  - **Carboxyl (-COOH)**.
  - **Amine (-NH₂)**.
  - **R-group (Side chain)**.
- **Proteins**:
  - **Primary structure**: Sequence of amino acids.
  - **Peptide bonds** link amino acids.

---

## 16.3 Summary**
- Life is built on **four main biomolecules**:
  1. **Carbohydrates** (Energy & structure).
  2. **Lipids** (Membranes & storage).
  3. **Nucleotides** (Genetics & energy).
  4. **Proteins** (Structure & function).
