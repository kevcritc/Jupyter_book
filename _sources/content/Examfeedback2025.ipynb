# 21. Exam Feedback from 2025 Exam Paper

## 21.1 Section A  
Section A comprised of a set of 5 Mark questions which generally require the students to address a few points, draw a diagram with particular elements included, or perform a calculation through the manipulation of an equation.

### 21.1.1 Question 1

The first question required you to describe the importance of miniturisation by clearly stating that it is making devices or machines smaller while performing the same or better. You would then state some advantages in your discussion 'weight' 'less material' 'higher functionality' etc would all have been suitable. Answers to this bit where mixed some just lacking clairty or just not stating any advantages at all.  Examples used by most where done well e.g. smart phone and computer.  I hoped to see more variety but this is probably due to the focus on the examples in the lectures.  Most scored well on this question.

### 21.1.2 Question 2 

This question was either done well or very badly with little inbetween.  The question was about scaling laws and used an example of bending a football through the air.  Recognising what parameters scale and which ones do not was the challenge.  Those who got this wrong generally thought that parameters such as speed scaled with size leading to the wrong answer in the first part.  Some found the scaling for the force and not the acceleration - they lost a mark.  The second part was to calculate the change with scale.  As long as the calculation was clear, marks were not lost if part i's answer was incorrect.  Many people did arrive at the correct answer. I hoped that it would be recognised that a smaller ball would increase the acceleration - bend would be stronger effect.  This is generally observed in sport cricket ball, table tennis, tennis ball. Getting a numerial answer correct approximately 35 $ \text{m} \text{s}^{-1} $.  

### 21.1.3 Question 3  

Most answered this well.  You needed to describe the three Bohr postulates and enhance it with diagrams and relevant equations to get all five marks.  


### 21.1.4 Question 4  

Almost everyone was able to answer this question.  You had to describe what wave-particle duality is.  Then describe an experiment such as double slit experiment making it clear when there is particle and wave behaviour being observed.  Some marks were lost just for not making one of these observations clear.  Two different experiments was accepted. A statement about how wave-particle duality and its use for the developemnt of quantum mechanics gave a final mark.

### 21.1.5 Question 5

Here you need to describe the shape of the CNT.  State the hexagonal arrangement of carbon atoms.  Mention $ \text {sp}^2 $ bonding.  Then mention two applications e.g. in sport equipment due to mechanical strength etc. and link to why they are good for the application.
Most obtain a majority of marks but missed the hybrid bonding.

### 21.1.6 Question 6

State what SPR is by describing the oscillation of electrons matching the frequency of the em source.  State that increasing size redshifts the SPR WL. State that changing the shape can give longer WL and/or additional SPR bands. Name an application and how the SPR property is useful.   Most obtained at least 3 marks, but typically missed the effects of  both size or shape.

### 21.1.7 Question 7  
Describe an amino acid structure. Describe formation of peptide. Assembly into alpha helix and beta sheets, then assembly into protein shape.  Describe or name a protein and it's function e.g.  hemoglobin is an iron rich protein which carries oxygen around the body - there are hundreds if not thousands of possible answers.  Most described the aminoacid but few described the primary, secondary and tertiary assemblies.  Most said something vague like proteins repair cells which was not great but I was generous as this is a physics module.

### 21.1.8 Question 8
Most answered the targeted drug delivery question well.  I was expecting something about active and passive targeting, something about lower dose, fewer side effects.  Some mention of how the targeting is achieved.  An example - was expecting something like the polymer conjugates but many discussed gold nanoparticle treatments which I accepted even though it is not really a 'drug' and certainly not a treatment used in the clinic.  

## 21.2 Section B
These were longer multipart questions which typically mix themes from the course in some way and try to test deeper understanding.

### 21.2.1 Question 1

This question started by asking for a diagram of energy levels for semiconductors and metals.  This is a very common question and I wa suprised that many did not answer it very well.  It is a diagram which is drawn many times in the lectures.

Then you were asked to calculate the size of the QD given the equation.  You had to read the peak position. Some chose to measure the upper and lower limits and take the difference which showed a lack of understanding of the system.  Others got tangled up in the units.  The equation needed rearranging and was given in the exam but some decided to add the Columb term as well to make it significantly more difficult than it needed to be.  Some struggled to rearrange the equation.  On the whole many struggled with the numerical solution which should be a easy mark for physics student with an A level in Maths.  Many answers were not in the nanoscale and with a basic critical check of the answer should have raised an alarm that something might be wrong.  The answer was 8.7 nm.

Next was to draw a wavefunction of a QD.  They should look like an atoms wavefunction. Mixed results here.  

The next bit was to stretch knowledge and if the above was not achieved then typically this was not answered.  I was looking for overlap of the wavefunctions.  If they overlap you will get a similar effect as we get with the formation of molecular orbitals and this is seen in quantum dot solids.  The effects you might guess are lowering of the bandgap - red shift- and possible a broadening of the peak.  Other effects such as energy transfer aggregation, and so on where accepted - anything reasonable.

Finally there was a question about environmental effects. Most people could discuss some advantages and disadvantages and potential for problems in disposal etc.

### 21.2.2 Question 2

You were expected to describe hybridisation of s and p orbitals in carbon and lead this to sp2 which is relevent for graphene. Two diagrams showing the orbitals and the graphene structure, respectively, obtained a mark each. Properties mechanical, electrical and thermal should then have been explained.  This question was mostly answered but some struggled with the diagrams and sp2 arguments.

Include two examples e.g. energy storage and electronics and what properties of the graphene are being used. Answered well.

The mixing in water was about hydrophobicity which most people got, but then the idea of coating was understoof by some and got vague.  Polymer modification of coating with a self-assembled monolayer was acceptable.

The final question was simple and only required numbers to be entered into the equation correctly to get $3.8 \times 10^{-4} $ J.  
Most did ok on this question.

### 21.2.3 Question 3
This question was answered poorly, perhaps due to time constraints. 

Diagram of a SAM with head group, alkane tail, and functional tail group was needed witha brief on what each does.

Diagram of $\text{H}_2\text{O}$ showing the lone pairs in sp orbitals.  Then diagrame showing the attracting through charge between the H and lone pairs. This was not done well.

Most could describe hydrophobic and hydrophilic but were vague on how to make a SAM that does this.

The numerical question was actually trivial.  You were given concentration on the surface in moles per unit area. Multiply by Avogadro's number to get molecules per area and then invert to get area of the molecule.  Change units from $ \text{cm}^2$ to $ \text{nm}^2$.  This was done poorly (by most).  The answer was 0.36 $ \text{nm}^2$. 

Know that the surface area of the sphere is $ 4 \pi r^2 $ and calculate the surface area.  Divide the surface area, by the area of a single molecule to get 144.

The order decreases because of the geometry means tight packing of the tails is impossible.
